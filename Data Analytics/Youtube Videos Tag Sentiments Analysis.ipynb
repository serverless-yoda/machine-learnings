# Perform sentiment analysis on Youtube Video Tags

```python
import pandas as pd
import numpy as np
import seaborn as sns
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
```

```python
# Read CSV and perform clean up of unnecessary characters
def sentiments():
    df = pd.read_csv('./data/youtube/USVideos.csv',error_bad_lines=False)
    #drop null fields
    df.dropna(inplace=True)
    video_all_tags = ' '.join(df['tags'])
    
    #remove all non-characters
    cleaned_text = re.sub(' +',' ',re.sub('[^a-zA-Z]',' ',video_all_tags))
    return cleaned_text      
```

```python
# generate cloudwords
def plotResult(data):
    plt.figure(figsize=(20,10))
    plt.imshow(data)
    plt.axis('off')
```

```python
# extract the sentiments
def generateWordCloud(data):
    return WordCloud(width=1000,height=200,stopwords=set(STOPWORDS)).generate(data)
    
```

```python
# display sentiments
word_data = generateWordCloud(sentiments())
plotResult(word_data)
```

# Check relation of Views and Like

```python
df = pd.read_csv('./data/youtube/USVideos.csv',error_bad_lines=False)
sns.regplot(data=df,x='likes',y='views')
plt.title('Views and Likes Regression')
```

# Check relation of Views and Dislike

```python
sns.regplot(data=df,x='dislikes',y='views')
plt.title('Views and DisLikes Regression')
```

# Check correlations between Likes,Dislikes and Views

```python
sns.heatmap(df[['views','likes','dislikes']].corr(),annot=True)
plt.title('Correlation of Views, Likes and Dislikes')
```
