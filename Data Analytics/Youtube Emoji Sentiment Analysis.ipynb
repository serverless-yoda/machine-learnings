```python
def emoji_list():
    df = pd.read_csv('./data/youtube/GBcomments.csv',error_bad_lines=False)   
    emojis = ''
    for item in df['comment_text']:        
        elist = [c for c in item if c in emoji.UNICODE_EMOJI['en']]       
        for emj in elist:
            emojis = emojis + emj
    return emojis
```
